```python
with open('cod.txt', encoding='utf8') as file:
    paf = file.read()
intab = "ауцльм"
outtab = "34*№%?"
trantab = str.maketrans(intab, outtab)
r = paf.translate(trantab)
```

```python
with open('cod.txt', 'w', encoding='utf8') as file:
    file.write(r)
```

```python
with open("cod.txt", encoding='utf8') as output_file:
    print(output_file.read())
```
