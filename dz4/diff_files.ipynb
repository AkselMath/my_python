```python
with open('vtoroiFail.txt', encoding = 'utf8') as file:
    planets = file.read()

intab = "\n"
outtab = " "
trantab = str.maketrans(intab, outtab)
planets = planets.translate(trantab)
planets
```

```python
y = 0
k = planets.index(' ')
a = [planets[0:k]]
while k < planets.count(' '):
    y = k + y + 1
    k = planets[y::].find(' ')
    if k == -1:
        break
    else:
        a.append(planets[y:y+k])
```

```python
trantab = str.maketrans("", "", " ,.!?")
n = 0
while n < len(a):
    a[n] = a[n].translate(trantab)
    if a[n] == '':
        a.pop(n)
    n += 1
```
